```python
import pandas as pd
```

```python
dev_df     = pd.read_csv('dataset/BC7-LitCovid-Dev.csv', sep=',')
test_gs_df = pd.read_csv('dataset/BC7-LitCovid-Test-GS.csv', sep=',')
test_df    = pd.read_csv('dataset/BC7-LitCovid-Test.csv', sep=',')
train_df   = pd.read_csv('dataset/BC7-LitCovid-Train.csv', sep=',')
```

```python
# Filling any NaNs with empty string ''

test_gs_df.fillna('', inplace=True)
train_df.fillna('', inplace=True)
dev_df.fillna('', inplace=True)

print(test_gs_df.isnull().sum(), '\n')
print(train_df.isnull().sum(), '\n')
print(dev_df.isnull().sum())
```

```python
# Dropping pmid, doi, journal, pub_type and authors columns

test_gs_df.drop(columns=['pmid', 'doi', 'journal', 'pub_type', 'authors'], inplace=True)
train_df.drop(columns=['pmid', 'doi', 'journal', 'pub_type', 'authors'], inplace=True)
dev_df.drop(columns=['pmid', 'doi', 'journal', 'pub_type', 'authors'], inplace=True)
```

```python
# Concatenating all column features into one column
# Adding a [CLS] token at the start of each row
# Adding a [SEP] token between each column concats

train_df['features'] = ('[CLS] ' + train_df['title'].astype(str) + 
                        '[SEP] ' + train_df['keywords'].astype(str) + 
                        '[SEP] ' + train_df['abstract'].astype(str))
train_df = train_df[['features', 'label']]
print(train_df)

test_gs_df['features'] = ('[CLS] ' + test_gs_df['title'].astype(str) + 
                        '[SEP] ' + test_gs_df['keywords'].astype(str) + 
                        '[SEP] ' + test_gs_df['abstract'].astype(str))
test_gs_df = test_gs_df[['features', 'label']]
print(test_df)

dev_df['features'] = ('[CLS] ' + dev_df['title'].astype(str) + 
                        '[SEP] ' + dev_df['keywords'].astype(str) + 
                        '[SEP] ' + dev_df['abstract'].astype(str))
dev_df = dev_df[['features', 'label']]
print(dev_df)
```

```python
test_gs_df.to_csv('dataset/cleaned/test_df.csv', sep=',')
train_df.to_csv('dataset/cleaned/train_df.csv', sep=',')
dev_df.to_csv('dataset/cleaned/validate_df.csv', sep=',')
```
